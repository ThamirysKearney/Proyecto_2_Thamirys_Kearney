## Creamos una copia del dataset original para aplicarle cambios importante como borrar algunas columnas, cambiar algunos datos,etc

### Borramos las columnas que no nos aporan mucha información de valor, en este caso borraremos las columnas partner_id,country_code,tags,borrower_genders y date

```python
## Borramos las columnas que no nos aporan mucha información de valor, en este caso borraremos las columnas partner_id,country_code
cols_to_drop = ['partner_id', 'country_code', 'tags', 'borrower_genders','funded_time','date','region','use']
for col in cols_to_drop:
    if col in kiva_loans_df.columns:
        kiva_loans_df =kiva_loans_df.drop(columns=col)

kiva_loans_df.info()
```

```python
#Convertimos las columnas de fecha a formato datetime
kiva_loans_df['posted_time'] = pd.to_datetime(kiva_loans_df['posted_time'],errors='coerce')
kiva_loans_df['disbursed_time'] = pd.to_datetime(kiva_loans_df['disbursed_time'],errors='coerce')

#Normalizamos a la fecha(elimnar la hora,estableciendola en medianoche) manteniendo el formato datetime
kiva_loans_df['posted_time'] = kiva_loans_df['posted_time'].dt.normalize()
kiva_loans_df['disbursed_time'] = kiva_loans_df['disbursed_time'].dt.normalize()
```

### Convertimos las columnas de object a string

```python
for col in kiva_loans_df.columns:
    if kiva_loans_df[col].dtype == 'object':
        kiva_loans_df[col] = kiva_loans_df[col].astype('string')
```

```python
kiva_loans_df.info()
```

### Nos aseguramos que los datos de tipo "str" no tengan espacios innecesarios y que estén escritas en minúscula

```python
def clean_text_columns(s):
    return s.astype('string').str.strip().str.lower()

for col in kiva_loans_df.columns:
    if kiva_loans_df[col].dtype == 'string':
        kiva_loans_df[col] = clean_text_columns(kiva_loans_df[col])
```

## Aqui realizamos la clasifiación de préstamos según la fecha de desembolso
### Creamos la columna loan_type para saber si un prestamo fue desembolsado antes o después de la fecha de publicación:
    - "pre_disbursed": el prestamo se desembolsó antes de publicarse(disbursed_time<posted_time)
    - "post_disbursed": el prestamo se desembolsó después de publicarse(disbursed_time>=posted_time)
    

```python
kiva_loans_df['loan_type'] = np.where(
    kiva_loans_df['disbursed_time'] < kiva_loans_df['posted_time'],
    'pre_disbursed',
    'post_disbursed'
)
```

## Verificamos la existencia de outliers (valores atípicos),respecto a las columnas funded_amount: y loan_amount

### Regla de Tukey es una forma sencilla de detectar valores atípicos(outliers) usando cuartiles. La idea es ver que tan lejos están los datos de la parte "normal" del conjunto.

    - Q1 = Valor donde termina el 25% más bajo de los datos
    - Q2 = Valor donde termina el 75% de los datos
    - IQR = Q3-Q1 -> mide la zona donde vive la mitad central de los datos

### Un dato se considera outliers si está demasiado lejos de la zona:

    - Límite inferior = Q1 - 1.5 x IQR
    - Limite superior = Q3 + 1.5 x IQR

### Todo lo que quede fuera de ese rango se marca como atípico.

```python
kiva_loans_df['loan_amount_category'].value_counts()
```

```python
categories = ["micro", "small", "medium", "large"]

for cat in categories:
  
  df_cat = kiva_loans_df[kiva_loans_df["loan_amount_category"] == cat]

  if df_cat.empty:
    print(f"\n{cat}: (sin datos)")
    continue
  
  q1 = df_cat["loan_amount"].quantile(0.25)
  q3 = df_cat["loan_amount"].quantile(0.75)

  iqr = q3 - q1
  # Limites para detectar outliers

  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr

  # Filtrar outliers

  outliers = df_cat[(df_cat["loan_amount"] < lower_bound) | (df_cat["loan_amount"] > upper_bound)]
  n_outliers = len(outliers)
  
  print(f"\nCategoría: {cat}")
  print(f"  Q1: {q1}")
  print(f"  Q3: {q3}")
  print(f"  IQR: {iqr}")
  print(f"  Outliers: {n_outliers}")
    
```
